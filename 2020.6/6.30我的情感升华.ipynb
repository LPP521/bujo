## 情感升华

这个命题绝对也不简单，我发现掌握同理心及共情力的始终也是少数，那么我的目的就是成为这少数人之一。

### 为什么我对家里人的行为极度恶心，又该怎么应对？

每次都以愤怒解决，看起来短时间内有效，但实际上只是暂时的。回想因愤怒而后悔做的事情，愤怒真的是降低智商且不理智的一种的情感宣泄。并不是每个人都能想到这个问题，既然我想到了，这就得想办法解决掉它。

首先，为什么会产生这样的心理？

我是在大家嘲讽中与家庭及学校被揍过程中长大的，现在，家庭开始对我和善，让我异常恶心。

控制自己的情感还需要正确的认识自己和他人，去找寻去求解这方面的答案，不然始终都是情绪的奴隶。共情力给了我启发：共情力需要换位思考，而我记得之前收藏了换位思考，能做到换位思考的前提，是要有能充分了解对方才能进行换位思考。那么怎样了解对方，如何知道对方的历史，叙述家常及过往。

先把事件记录下来，愤怒时转移注意力，再将“我”置换为第三人称，查找是否存在与“我”一样的反家庭社会群体，探究这一个现象。

### 提升自我的教养

教養是什麼呢？

它是心中的分寸，是言行的適度；是對他人的尊重，是平等與包容。

> 可是被你傷害的那個人，卻永遠不會忘記你。他絕不會記住你的優點，而是記住你對他的傷害。- 卡耐基

> 好的教養，讓人賞心悅目、如沐春風，一個人缺乏教養，則讓人心生厭惡。- 佚名

这句话让我想到我的初恋，再联想到我学生时期目中无人失去了多少朋友，青春期的叛逆让人心生厌恶，在社会上遭受了多少的毒打，才知道造成了多少遗憾啊，我已决定改头换面！社交的积累，习惯的养成，看来教养也绝非易事，必须认真对待了。


#### 伪化


> [怎样的言行举止可以反映一个人的教养和内涵？](https://www.zhihu.com/question/27871104/answer/45664502)


1. 搞清自己与对方的身份，方位
1. 关注对方所说的内容、喜好、烦恼
1. 尊重他的感觉、观点，哪怕那感觉在你看来很蠢
1. 尊重他的习惯，哪怕你并不会那样做，哪怕你觉得那样很麻烦
1. 尊重他的时间，哪怕你确切知道他很闲，也要假定他很忙
1. 尊重人家的物品，哪怕只是你觉得很便宜的东西
1. 尊重人家的金钱，尤其是花在自己身上，哪怕没多少
1. 说话主动承担责任


#### 共情

> 参考 [简单心理uni-如何培养共情能力？](https://www.zhihu.com/question/24440657/answer/463431187)

将自己转变为心理咨询师

> 对非言语行为的反馈（Reflecting nonverbal behavior）

对方向自己诉苦时，观察对方行为，说出自己对对方行为的直觉感受。

> 对深层情感的反馈（Reflecting deeper feelings）

表示同情对方所做的一切挫败感，并说明原因

> 指出有冲突性的感觉和想法（Pointing out conflictual feelings and thoughts）

指出矛盾点，询问新事物有哪些吸引他的特质

> 运用视觉意象（Using visual imagery）

告诉自己正在模拟环境他的环境，得出一个共同参考的结论

> 运用类比（Using analogies）

根据他的描述做类比得出相似的结论

> 有目标性的自我暴露（Using targeted self-disclosure）

揭露自己相似的经历，与他所处同一立场

> 运用媒体来反馈（Reflecting media）

以名人事迹、书、电影给他带来希望

> 以一种可感触到的方式去给予反应（Reflecting tactile responses）

作为他的一面镜子，“揪心”“被绞般疼痛”我想这是你所感受到的。先死后活，经过多方尝试再去不断调整

### 心态调整

看了这两篇话题，原来他们也有和我有这般的故事，甚至他们比我还不幸。

* [知乎-作为一名成年人，被父母打了，如何调整心态？](https://www.zhihu.com/question/40984913/answer/555662959)
* [知乎-面对从童年持续至今的家庭暴力如何调整心态走出阴影？](https://www.zhihu.com/question/31277152/answer/51423535)

观点汇总：

* 经济独立远离他们
* 理性看待，经验主义使得循环施暴的人，且施暴者本身就不具有理性，根本不会去想经验外的解决方式
* 当它作为文化的一种特质被广泛接受后，共同的命运体的意识，本身有疗愈的功能
* 当它过于严重时，应找专业心理咨询师 
* 或将自己的事迹总结传播出去，用思想来帮助更多的人
