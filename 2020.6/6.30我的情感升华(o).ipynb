## 情感升华

这个命题绝对也不简单，我发现掌握同理心及共情力的始终也是少数，那么我的目的就是成为这少数人之一。

### 为什么我对家里人的行为极度恶心，又该怎么应对？

每次都以愤怒解决，看起来短时间内有效，但实际上只是暂时的。回想因愤怒而后悔做的事情，愤怒真的是降低智商且不理智的一种的情感宣泄。并不是每个人都能想到这个问题，既然我想到了，这就得想办法解决掉它。

首先，为什么会产生这样的心理？

我是在大家嘲讽中与家庭及学校被揍过程中长大的，现在，家庭开始对我和善，让我异常恶心。

 我的原因是：
 
 1. 过度关心的情感越界

控制自己的情感还需要正确的认识自己和他人，去找寻去求解这方面的答案，不然始终都是情绪的奴隶。共情力给了我启发：共情力需要换位思考，而我记得之前收藏了换位思考，能做到换位思考的前提，是要有能充分了解对方才能进行换位思考。那么怎样了解对方，如何知道对方的历史，叙述家常及过往。

先把事件记录下来，愤怒时转移注意力，再将“我”置换为第三人称，查找是否存在与“我”一样的反家庭社会群体，探究这一个现象。

### 我为何而抱怨？

[常爱说「废话」及爱抱怨的人，都是抱着什么心态的？](https://www.zhihu.com/question/22811035)

解决问题并不是抱怨的初衷，对于当时的我而言只是在“陈述事实”，想要得到情感上的理解和认同。那么，这么看其实抱怨其实挺正常的，现在该怎样去应对“我”呢？

“我”换一种抱怨的表达方式：我有一个朋友他感觉blablabla，暂缓朋友对抱怨的厌恶，并提出解决方案。

* 认同“我”的感觉
* 对“我”的情绪表示理解
* 给予“我”最大情感支持

结合我自己的以往与我爸、我爷爷（文革时期），这篇[知乎日报-爱抱怨的人，都是抱着什么心态的？](https://www.zhihu.com/question/22811035/answer/31413064)的几点也说得在理。

* 喜欢抱怨的人一般都是心眼特别实在的人，他们不擅长于察颜观色
* 喜欢抱怨的人，承受生活压力的心理能力相对比较弱
* 爱抱怨的人，并没有什么心理问题，抱怨只是行为特色

抱怨产生的成因：至少我家内，他们心眼实在也不太会与人交流，想到什么说什么，也没有忌讳，他们只想需要有人倾听，减缓他们心理压力。那么，该怎么解决：通过情感理解的方式，让他们知道他们所抱怨的内容运作原理，尝试以“不知不觉”的方式引导他用其他方式付诸行动。

> 摘自 [知乎-倾诉与抱怨有区别吗？](https://www.zhihu.com/question/24573922/answer/32048982)

倾诉偏向客观陈述事实，倾诉者的心态是端正的没有强烈的喜好偏向。倾诉者有意向听取倾听者的意见。抱怨偏向主观感受的表达，过分强调除了自身以外因素对事情的消极影响度，而不自省。最关键的区别是抱怨者没有准备听取倾听者的反馈，有意向和决心去改变现状，却只想发泄自身消极情绪而已。

### 一路走来的负面偏激

在一次过年中，亲戚对我说：“你在外打工有段时间了，一起相互加下微信，也好对他们有个照应”。我当时在想：现在内卷有那么严重吗？我靠，连你们这些整天无所事事、跳广场舞的大妈都觉醒了...合着我像一只被实验的小白鼠，给你们探路避坑用的。但我现在就一底层普工，怎么照应你们呢，也许等我以后吧。但随即我对亲戚这一行为矛盾似乎不太符合理性利益需求的背后，产生了很大不解的心结及疑问。

在朋友的开导下，你看到路边的乞丐给予施舍，是因为利益吗？后来，我才想明白其实他们是因我过往遭遇产生的同情想着帮来我。因自己遭遇的不幸对比身处小资阶级游乐的亲戚的心生憎恨，本来朋友就很少了，当时要是为这事表露态度的话，很可能再又把“不离不弃”的亲戚们推远了，我真的就成孤家寡人了。

#### 共情

> 参考 [简单心理uni-如何培养共情能力？](https://www.zhihu.com/question/24440657/answer/463431187)

将自己转变为心理咨询师

> 对非言语行为的反馈（Reflecting nonverbal behavior）

对方向自己诉苦时，观察对方行为，说出自己对对方行为的直觉感受。

> 对深层情感的反馈（Reflecting deeper feelings）

表示同情对方所做的一切挫败感，并说明原因

> 指出有冲突性的感觉和想法（Pointing out conflictual feelings and thoughts）

指出矛盾点，询问新事物有哪些吸引他的特质

> 运用视觉意象（Using visual imagery）

告诉自己正在模拟环境他的环境，得出一个共同参考的结论

> 运用类比（Using analogies）

根据他的描述做类比得出相似的结论

> 有目标性的自我暴露（Using targeted self-disclosure）

揭露自己相似的经历，与他所处同一立场

> 运用媒体来反馈（Reflecting media）

以名人事迹、书、电影给他带来希望

> 以一种可感触到的方式去给予反应（Reflecting tactile responses）

作为他的一面镜子，“揪心”“被绞般疼痛”我想这是你所感受到的。先死后活，经过多方尝试再去不断调整
