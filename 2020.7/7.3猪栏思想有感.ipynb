
### 初见猪栏


> 我从来不把安逸和快乐看作是生活目的本身, 对这种伦理基础, 我称之为’猪栏的理想’。— 爱因斯坦

安逸和快乐看作是生活目的本身，难道不是吗？这对于我认知有些颠覆，刚开始一下始终没法理解。他是怎么会得出这个结论的呢？随即开始在网上搜寻他说这段话的历史与讨论。果然是完整一部书的摘抄...去除整体，留下部分，歧义很大的啊！


> 摘自 爱因斯坦 《我的世界观》
> 要追究一个人自己或一切生物生存的意义或目的，从客观的观点看来，我总觉得是愚蠢可笑的。可是每个人都有一定的理想，这种理想决定着他的努力和判断的方向。就在这个意义上，我从来不把安逸和快乐看作是生活目的本身——这种伦理基础，我叫他猪栏的理想。照亮我的道路，并且不断地给我新的勇气去愉快地正视生活的理想，是善、美和真。要是没有志同道合者之间的亲切感情，要不是全神贯注于客观世界——那个在科学与艺术工作领域永远达不到的对象，那么在我看来，生活就会是空虚的。人们所努力追求的庸俗的目标——财产、虚荣、奢侈的生活——我总觉得都是可鄙的。　　


看来读名言还是不能只看名言啊，名言前后都是有铺垫的。他（名人）所处的环境，得出的结论（名言），放在我身上可未必适用，因为我还是要混饭吃的人，现今远没有财务自由。我的一些疑问来了：


我的生活的理想又是什么？

* 跨越当前现状，往上爬，成为能有生产资料的小资产阶级（大方向）
* 走向职场管理层，业余发展一门语言
* 学会去贡献自己的共情力及同理心（一生）


究竟怎样才算是善呢？

一个思考的事例：

* [什么才是真正的善良？ - 网易槽值的回答 - 知乎](https://www.zhihu.com/question/34342275/answer/281095955)

我的想法：那么什么是善呢？是共情、同理心去爱吗？是为一个不幸的遭遇，不合理运作，提出一个解决方案吗？

> [Cody Lee-什么才是真正的善良？-知乎](https://www.zhihu.com/question/23167335/answer/87460518)

> 亚当斯密在《道德情操论》里对克己有这么一段类似的描述，“一个对身体疼痛毫无感觉得人，即使以最完美无暇的耐性与镇定忍受了酷刑折磨，也不值得任何人为其鼓掌。一个天生异常不害怕死亡的人，即使在最可怕的危险环绕中保持他的冷静与沉着，也没有资格为此而声称他自己有什么了不起的优点。”回到善与恶的话题，即使真有上帝，作为一个人类我们应该非常自豪地认为我们的善比上帝的善更伟大。上帝所谓的善是来自与他本身完美的性质，不需要努力和克制；而人类的善，则是我们自己修来的恩典，完全来自我们自己的不懈努力。我们没有必要像基督徒一样觉得我们人性中的恶是什么不可饶恕的罪过。正应为我们有作恶的倾向，我们的善良才有价值；正式因为恶的存在，我们才有一个机会证明人类有不任由欲望摆布的尊严。

就算因某些不可抗被动条件不得不做出的生存选择（谋略），即便这样也能保持初心，一直向往那个方向行动吧。
